Encoding into latin 1

UTF-8 is the default rulebook that works for most text.

Sometimes, we have text with special characters or accents that don't fit well with the default rules.

'latin-1' is another rulebook that can handle more types of characters and symbols, its mainly used for these purposes
1. Character Compatibility:
'latin-1' works for a wide range of characters, like those from Western European languages and special symbols, without causing errors.
2. Preservation of Data:
When data has tricky characters not suited for UTF-8 (like special symbols or accents), 'latin-1' keeps the data intact when reading it.
3. Legacy Data:
Sometimes, older systems or sources use 'latin-1' encoding, so it's needed to read and understand the data accurately.
4. Compatibility with Software:
In some cases, 'latin-1' is chosen to work with specific software or tools that expect or demand it.

```python
#I simply wrote a prompt for chatGPT to show me in a grapgh population of heights
#from which i am taking a sample which is in red circle round shape while making sure that both have
#same mean, standard deviation and variance

#***************************************************************************
import numpy as np
import matplotlib.pyplot as plt

# Generate population heights data
np.random.seed(0)  # Setting a seed for reproducibility
population_heights = np.random.normal(170, 10, 1000)

# Create a scatter plot of population heights
plt.figure(figsize=(8, 4))
plt.scatter(range(len(population_heights)), population_heights, s=20, alpha=0.5)
plt.title('Population of Heights')
plt.xlabel('Individuals')
plt.ylabel('Height (cm)')

# Create a sample from the population
sample = np.random.choice(population_heights, size=30)

# Create a scatter plot for the sample
plt.figure(figsize=(8, 4))
plt.scatter(range(len(population_heights)), population_heights, s=10, alpha=0.5) # alha is for transparency it can make colors look less bright
plt.scatter(range(len(sample)), sample, c='red', marker='o', s=50, label='Sample Data', alpha=0.5)
plt.title('Population and Sample Heights')
plt.xlabel('Individuals')
plt.ylabel('Height (cm)')
plt.legend()

# Calculate and display the statistics for population and sample
population_mean = np.mean(population_heights)
sample_mean = np.mean(sample)
population_std = np.std(population_heights, ddof=1)  # Use ddof=1 degree of freedom for sample standard deviation
sample_std = np.std(sample, ddof=1)
population_var = np.var(population_heights, ddof=1)
sample_var = np.var(sample, ddof=1)

print(f"Population Mean: {population_mean:.2f}") #here .2f means it will show upto two digits after point
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Population Standard Deviation: {population_std:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Population Variance: {population_var:.2f}")
print(f"Sample Variance: {sample_var:.2f}")

plt.show()

#******************************************************************************
```

```python
# again i used chatGPT to visualize and help in learning this concept
# Creating a dataset to help understand this concept,
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    'Income': [45000, 55000, 60000, 75000, 80000, 90000, 95000, 100000, 105000, 110000],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Education Level': ['High School', 'Bachelor', 'Master', 'Ph.D', 'Bachelor', 'Master', 'Ph.D', 'Bachelor', 'Master', 'Ph.D'],
}

# Creating a DataFrame
df = pd.DataFrame(data)

#***********************************************************
# Defining varibales ml

target_variable = 'Income' # Dependent Variable
feature_variables = ['Age', 'Gender', 'Education Level'] # Independent Variables

# we are saying that income depends on age gender and education level

categorical_variable = 'Gender' # Categorical Variable
continuous_variable = 'Age' # Continuous Variable


#***********************************************************
#1st grapgh
# Plotting Continuous Variable (Age)
plt.figure(figsize=(8, 4))
plt.scatter(df[continuous_variable], df[target_variable])

# here df[continuous_variable] represents the data for the x-axis
# continuous_variable is the variable chosen for the x-axis, and it's typically considered the independent variable.


plt.xlabel(continuous_variable)
plt.ylabel(target_variable)
plt.title(f'Scatter Plot of {continuous_variable} vs. {target_variable}')
plt.show()

#this code is very straight forward and easy to understand
#***************************************************************
# 2nd graph

#plotting categorical variable
#percentages distribution is an excellet method for analysing categorical data

# Calculate income totals for each gender
gender_income_totals = df.groupby('Gender')['Income'].sum()

# Calculate the percentage of income for each gender
gender_income_percentage = (gender_income_totals / gender_income_totals.sum()) * 100


#total income for each gender category using groupby and sum.
#percentage of income for each gender
# for showing percentage  a pie chart to visualize the income distribution by gender, showing the percentage for each category

# Pie Plot the percentage distribution
#Question why Am I using a pie plot in this example?

plt.figure(figsize=(4, 4))
# this code is giving me the size of the plot

gender_income_percentage.plot(kind='pie', autopct='%1.1f%%')

#whats happening in this code?
# autopct='%1.1f%%': This part specifies how the percentage labels should be formatted and displayed on the pie chart
#   autopct stands for "auto percentage."
#   '%1.1f%%' is a format string that determines how the percentages are shown
#   %1.1f: This part means to display the percentage with one digit before the decimal point and one digit after. For example, 25.5%.
#   %%: The double percent sign is used to display the percentage symbol (%)
#  Since % is a special character, we need to use %% to display it as a literal percentage symbol in the chart.


plt.title('Income Distribution by Gender')
plt.ylabel('')  # Remove y-axis label because i am making a pie chart
#Question why have I left ylabel empty and no axis are defined either????
plt.show()
```

```python
#lets see this with help of example
# 1. Unimodal (Normal Distribution)
mean = 50
std_deviation = 10
unimodal_data = np.random.normal(mean, std_deviation, 100)  # Generate 100 data points
###In this code we will get 100 data points which have mean 50, standard deviation 10
#this means all data will be clustered around mean to give us the unimode (single peak data)
# create data frame from these data points
df_unimodal = pd.DataFrame({'Value': unimodal_data})

#*********************************************************************************

# 2. Bimodal
bimodal_data = np.concatenate((np.random.normal(25, 2, 50), np.random.normal(75, 5, 50)))  # Generate 100 data points

# now this code is simply adding two data points by concatenate fuction and random numbers in a bell shaped normal distribution
# mean is 25, sd 2 and genrate 50 data points + mean 75, sd 5 and 50 data points
# the resultant is a bi-modal data with two peaks.
# create data frame from these data points
df_bimodal = pd.DataFrame({'Value': bimodal_data})

#*********************************************************************************
# 3. Positively Skewed
skewed_data = np.random.exponential(10, 100)  # Generate 100 data points, exponential distribution is positively skewed
#skewness in a data simply means that it is not a bell shaped curve
# rather a shape with peak on right side or left side see the graph
df_skewed = pd.DataFrame({'Value': skewed_data})

#*********************************************************************************
# 4. Normal Distribution (Approximate) The bell shaped curve
mean = 60
std_deviation = 10
normal_data = np.random.normal(mean, std_deviation, 100)  # Generate 100 data points
df_normal = pd.DataFrame({'Value': normal_data})





#*********************************************************************************
# Plotting histograms as example to show
plt.figure(figsize=(12, 6))
# this code is simply giving the size of figure where 12 is width and 6 is length
#this code will determine the size of my graph


# Unimodal (Normal Distribution)
plt.subplot(221)
plt.hist(df_unimodal['Value'], bins=20, color='blue', alpha=0.7)
#"alpha"  controls the transparency or opacity
# "bins"  determines the number of intervals or classes
plt.title('Unimodal (Normal Distribution)')

# Bimodal
plt.subplot(222)
plt.hist(df_bimodal['Value'], bins=20, color='green', alpha=0.7)
plt.title('Bimodal')

# Positively Skewed
plt.subplot(223)
plt.hist(df_skewed['Value'], bins=20, color='red', alpha=0.7)
plt.title('Positively Skewed')

# Normal Distribution (Approximate)
plt.subplot(224)
plt.hist(df_normal['Value'], bins=20, color='purple', alpha=0.7)
plt.title('Normal Distribution (Approximate)')

plt.tight_layout()
plt.show()
```
